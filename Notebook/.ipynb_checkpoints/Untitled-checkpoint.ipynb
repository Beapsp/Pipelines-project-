### A continuación voy a crearme un dataframe con la información obtenida de la API elegida

### Compruebo que valores de la columna "Food" de mi df inicial y los valores de la columna "name" del df obtenido de la API coinden, por tanto puedo unir ambos dataframes por dichas columnas.

```python
mergeado = df.merge(df_api,left_on="Food", right_on="name", suffixes=["_izq", "_dcha"] )
```

```python
mergeado.head()
```
